```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

Air quality_no2 Dataset loading....

```python
air_quality = pd.read_csv('/content/air_quality_no2.csv')
```

```python
air_quality.head()
```

I want a quick visual check of the data.

```python
air_quality.plot()
```

```python
plt.show()
```

 I want to plot only the columns of the data table with the data from Paris.

```python
air_quality['station_paris'].plot()
```

Note

>To plot a specific column, use a selection method from the subset data tutorial in combination with the plot() method. Hence, the plot() method works on both Series and DataFrame.



I want to visually compare the
𝑁⁢𝑂2
 values measured in London versus Paris.

```python
air_quality.plot.scatter(x = "station_london", y = "station_paris", alpha = 0.5)
```

Note


**What is Scatter?**

>It is used to see if there is a relationship or correlation between two variables

**Comaparison**

alpha -> 1.0
> Every dot is solid
alpha -> 0.1
>Every dot is faint
alpha -> 0.5
>A middle ground works well for median sized datasets

```python
air_quality.plot.box()
```

Note

>In many development environments such as IPython and Jupyter Notebook, use the TAB button to get an overview of the available methods, for example air_quality.plot. + TAB.

I want each of the columns in a separate subplot.

```python
axs = air_quality.plot.area(figsize = (12, 4), subplots = True)
```

I want to further customize, extend or save the resulting plot.

```python
fig, axs = plt.subplots(figsize=(12, 4))
```

```python
air_quality.plot.area(ax=axs)
```

```python
axs.set_ylabel("NO$_2$ concentration")
```

```python
fig.savefig("no2_concentrations.png")
```

```python
plt.show()
```

Note

fig, axs = plt.subplots(figsize=(12, 4)) -->       
 Create an empty Matplotlib Figure and Axes

air_quality.plot.area(ax=axs)    ---->                   
 Use pandas to put the area plot on the prepared Figure/Axes

 axs.set_ylabel("NO$_2$ concentration")---->          
Do any Matplotlib customization you like

fig.savefig("no2_concentrations.png")    ----->
Save the Figure/Axes using the existing Matplotlib method.

plt.show() ---> Display the plot

**Remember**


>The .plot.* methods are applicable on both Series and DataFrames.
By default, each of the columns is plotted as a different element (line, boxplot, …).Any plot created by pandas is a Matplotlib object.
